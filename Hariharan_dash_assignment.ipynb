```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
```

```python
# Uses an External Stylesheet
# Use a css file from your GitHub Pages site 
external_stylesheets = ['https://hariharan-usf.github.io/data/learn.css']

# Creates the app to instantiate the content for the Dashboard and use the external_stylesheets
app = dash.Dash(__name__, external_stylesheets=external_stylesheets)
```

```python
# Use a csv dataset from a repository in your GitHub account. Use the Raw URL to expose the data to the Python dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Hariharan-USF/data/master/telcochurn')
#df2 = pd.read_csv('https://raw.githubusercontent.com/USFMumaAnalyticsTeam/data/master/usa-agricultural-exports-2011.csv')
```

```python
# Add content to the app layout
# Begin all content DIV
app.layout=html.Div([
    # Add your HTML tags to the content - notice a comma is added between HTML elements
    html.H1('Data Visualization for Telecom churn analysis using user data'),
    html.Div([
        html.P('Churn vs User demographics ...'),
    ]),
    # Begin of DIV surrounding both Tables
    html.Div([
    # Begin of First Table
    html.Table(style={'width':'100%'},
               # Begin of Table children
               children=[
                   #######################################################################
                   # Begin of First Tr
                 html.Tr(
                     #Begin Tr children
                     children=[
                         # Begin Th
                         
                         html.Th(style={'width':'50%'},
                             # Begin Th children
                             children=[
                                 html.H3('Churn vs gender')
                             # End of Th children   
                             ]
                         
                         # End of Th - Notice a comma is placed here to separate the next Th
                         ),
                         # Begin of Th
                         html.Th(style={'width':'50%'},
                             # Begin of Th children
                             children=[
                                 html.H3('Churn vs partner')
                             # End of Th children    
                             ]
                         
                         # End of Th
                         )
                         
                     # End of Tr children    
                     ]
                 # End of First Tr - Notice a comma is placed here to separate the next Tr
                 ),
                 #########################################################################
                 # Begin of Second Tr
                 html.Tr(
                     #Begin Tr children
                     children=[
                         # Begin Td
                         html.Td(
                             # Begin Td children
                             children=[
                                    # Display bar graph
                                    dcc.Graph(
                                    id='example-graph',
                                    figure={
                                        'data': [
                                            {'x': ["Male","Female"], 'y': [df[(df.gender == "Male") & (df.churn == "Yes")].shape[0],df[(df.gender == "Female") & (df.churn == "Yes")].shape[0]], 'type': 'bar', 'name': 'Churn_yes'},
                                            {'x': ["Male","Female"], 'y': [df[(df.gender == "Male") & (df.churn == "No")].shape[0],df[(df.gender == "Female") & (df.churn == "No")].shape[0]], 'type': 'bar', 'name': 'Churn_no'},
                                        ],
                                        'layout': {
                                            'title': 'Churn vs Gender'
                                        }
                                    }
                                # End of Chart 1
                                )
                              # End of Td children   
                             ]
                         
                         # End of Td - Notice a comma is placed here to separate the next Th
                         ),
                         # Being of Td
                         html.Td(
                             # Begin of Td children
                             children=[
                                 # Display plot graph use data from dataframe df
                                 dcc.Graph(
                                    id='life-exp-vs-gdp',
                                figure={
                                    'data': [
                                            {'x': ["Partner_yes","Partner_no"], 'y': [df[(df.partner == "Yes") & (df.churn == "Yes")].shape[0],df[(df.partner == "No") & (df.churn == "Yes")].shape[0]], 'type': 'bar', 'name': 'Churn_yes'},
                                            {'x': ["Partner_yes","Partner_no"], 'y': [df[(df.partner == "Yes") & (df.churn == "No")].shape[0],df[(df.partner == "No") & (df.churn == "No")].shape[0]], 'type': 'bar', 'name': 'Churn_no'},
                                        ],
                                        'layout': {
                                            'title': 'Churn vs Partner'
                                        }
                                    }
                                 )
                           
                             # End of Td children    
                             ]
                         # End of Td
                         )
                     # End of Tr children    
                     ]
                 # End of Tr
                 )     
                #########################################################################                   
               #End of Table Children    
               ]
              # End of First Table - Notice a comma is placed here to separate the next Table
              ),
    
    #########################################################################################   
    # Begin of Second Table
    html.Table(style={'width':'100%'},
               # Begin of Table children
               children=[
                   #######################################################################
                   # Begin of First Tr
                 html.Tr(
                     #Begin Tr children
                     children=[
                         # Begin Th
                         
                         html.Th(style={'width':'50%'},
                             # Begin Th children
                             children=[
                                 html.H3('Churn vs Dependents')
                             # End of Th children   
                             ]
                         
                         # End of Th - Notice a comma is placed here to separate the next Th
                         ),
                         # Begin of Th
                         html.Th(style={'width':'50%'},
                             # Begin of Th children
                             children=[
                                 html.H3('Churn vs Senior Citizen')
                             # End of Th children    
                             ]
                         
                         # End of Th
                         )
                         
                     # End of Tr children    
                     ]
                 # End of First Tr - Notice a comma is placed here to separate the next Tr
                 ),
                 #########################################################################
                 # Begin of Second Tr
                 html.Tr(
                     #Begin Tr children
                     children=[
                         # Begin Td
                         html.Td(
                             # Begin Td children
                             children=[
                                    # Display bar graph
                                    dcc.Graph(
                                    id='example-graph1',
                                    figure={
                                        'data': [
                                            {'x': ["Dependents_yes","Dependents_no"], 'y': [df[(df.dependents == "Yes") & (df.churn == "Yes")].shape[0],df[(df.dependents == "No") & (df.churn == "Yes")].shape[0]], 'type': 'bar', 'name': 'Churn_yes'},
                                            {'x': ["Dependents_yes","Dependents_no"], 'y': [df[(df.dependents == "Yes") & (df.churn == "No")].shape[0],df[(df.dependents == "No") & (df.churn == "No")].shape[0]], 'type': 'bar', 'name': 'Churn_no'},
                                        ],
                                        'layout': {
                                            'title': 'Churn vs Dependents'
                                        }
                                    }
                                # End of Chart 1
                                )
                              # End of Td children   
                             ]
                         
                         # End of Td - Notice a comma is placed here to separate the next Th
                         ),
                         # Being of Td
                         html.Td(
                             # Begin of Td children
                             children=[
                                 # Display plot graph use data from dataframe df
                                 dcc.Graph(
                                    id='life-exp-vs-gdp1',
                                figure={
                                    'data': [
                                            {'x': ["SeniorCitizen_yes","SeniorCitizen_no"], 'y': [df[(df.seniorcitizen == 1) & (df.churn == "Yes")].shape[0],df[(df.seniorcitizen == 0) & (df.churn == "Yes")].shape[0]], 'type': 'bar', 'name': 'Churn_yes'},
                                            {'x': ["SeniorCitizen_yes","SeniorCitizen_no"], 'y': [df[(df.seniorcitizen == 1) & (df.churn == "No")].shape[0],df[(df.seniorcitizen == 0) & (df.churn == "No")].shape[0]], 'type': 'bar', 'name': 'Churn_no'},
                                        ],
                                        'layout': {
                                            'title': 'Churn vs Senior Citizen'
                                        }
                                    }
                                 )
                           
                             # End of Td children    
                             ]
                         # End of Td
                         )
                     # End of Tr children    
                     ]
                 # End of Tr
                 )     
                #########################################################################                   
               #End of Table Children    
               ]
              # End of First Table - Notice a comma is placed here to separate the next Table
              ),
    
    # End of DIV surrounding both Tables
    ]),
               
# End of all content DIV
])
```
